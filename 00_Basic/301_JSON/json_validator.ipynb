<h1>JSON Validation</h1>

```python
import json 
from jsonschema import validate, ValidationError, SchemaError
```

```python
address_schema = json.load(open('resources/address_schema.json'))
address_data = json.load(open('resources/address.json'))
```

```python
try:
    validate(address_data, address_schema, format_checker=Draft7Validator.FORMAT_CHECKER)
    print("JSON data is valid against the schema.")
except ValidationError as e:
    print(f"Validation error: {e.message}")
    exit(1)
except SchemaError as e:
    print(f"Schema error: {e.message}")
    exit(1)
```
