```python
s_hello[0] = "h"  # This will raise an error because strings are immutable
```

```python
s_hello = "Hello, World!"
for i in range(len(s_hello)):
    print(s_hello[i], end="   ")

print("\n")

for i in range(len(s_hello)):
    print(i, end="   ")

print("\n")
for i in range(-1, -len(s_hello), -1):
    print(i, end="   ")
```
