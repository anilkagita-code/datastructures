## Insertion Sort

### The paradime is called Recursion

##### Explanation

##### Efficiency 

T(n) = 1 + 2 + ... + (n-1) + n ==> n(n+1)/2 ==> n**2/2 + n/2

##### O(n**2)

```python
def insertion_sort(seq):
    
    # Reading the first element from the Unsorted sequence
    for sorted_end_idx in range(len(seq)):
        
        # Move the first element from the unsorted sequence and 
        # place it into the right position in the sorted sequence
        cur_idx = sorted_end_idx
        
        while cur_idx > 0 and seq[cur_idx] < seq[cur_idx-1]:
            seq[cur_idx], seq[cur_idx-1] = seq[cur_idx-1], seq[cur_idx]
            cur_idx -= 1
            
    return seq
```

```python
testlist = [74, 32, 89, 55, 21, 64]
insertion_sort(testlist)
```

##### Note: 
- In spite of both having the as worst case efficiency of O(n**2)
- Insertion sort can be better if the list is sorted, since it does only one operation to move the element from unsorted sequence to the sorted sequence
- Where are Selection sort runs through the entire list each time to check for least element and then swaps it with the current position



## Insertion Sort Recursive

```python
def insertion_sort_rec(seq):
    print(f'In insertion_sort_rec:  seq: {seq}, len(seq): {len(seq)}')
    rec_split(seq, len(seq))
    return seq
```

```python
def rec_split(seq, seq_len):
    print(f'In rec_split: seq: {seq}, seq_len: {seq_len} ')
    if seq_len > 1:
        rec_split(seq, seq_len-1)
        move_element_sorted_seq(seq, seq_len-1)
```

```python
def move_element_sorted_seq(seq, seq_len):
    print(f'In move_element_sorted_seq: seq: {seq}, seq_len: {seq_len} ')
    cur_idx = seq_len
    while cur_idx > 0 and seq[cur_idx] < seq[cur_idx-1]:
        seq[cur_idx], seq[cur_idx-1] = seq[cur_idx-1], seq[cur_idx]
        print(f'while move_element_sorted_seq: seq: {seq}, seq_len: {seq_len} ')
        cur_idx -= 1
```

```python
testlist = [74, 32, 89, 55, 21, 64]
insertion_sort_rec(testlist)
```

##### Efficiency for recursive Insertion Sort

T(n): time to run insertion sort on length n
- Time T(n-1) to sort slice sequence seq[0: n-1]
- n - 1 steps to insert seq[n-1] in sorted slice

T(n) = (n-1) + T(n-1) ==> (n-1) + (n-2) + T(n-2)

T(n) = (n-1) + (n-2)... + 1 ==> n(n-1)/2 ==> n**2/2 + n/2

##### O(n**2)
