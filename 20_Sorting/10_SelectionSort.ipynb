## Selection Sort

### The paradime is called Recursion

##### Explanation

```python
def selection_sort(seq):
    
    # Scanning the elemenets for 0:len, 1:len, ...
    for start_idx in range(len(seq)):
        
        # find the min value in the given list
        min_idx = start_idx
        for i in range(start_idx, len(seq)):
            if seq[i] < seq[min_idx]:
                min_idx = i
                
        # Exchange the values
        (seq[start_idx], seq[min_idx]) = (seq[min_idx], seq[start_idx])
    
    return seq
```

```python
testlist = [74, 32, 89, 55, 21, 64]
selection_sort(testlist)
```

##### Efficiency 

T(n) = n + (n-1) + (n-3) + ... + 1 ==> n(n+1)/2 ==> n**2/2 + n/2

##### O(n**2)
